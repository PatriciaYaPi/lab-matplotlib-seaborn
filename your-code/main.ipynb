#### Plot both $y=x^2$ and $y=e^x$ in the same plot using normal and logarithmic scale.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here

"""
.set_xscale(value)
value: {"linear", "log", "symlog", "logit", ...} 
"""

"""
plt.plot(x,
         x**2,
         linestyle='--',
         linewidth=3,
         color='#9baee4', #HEX colour
         #markersize=12,
         #marker='o'
        ) 
"""

#Linear scale:
##############
plt.plot(x, x**2, 'b--', x, np.exp(x),'g-')
plt.title('Linear Scale')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.xscale('linear') #default value
plt.yscale('linear') 

plt.ylim(0,1000)

plt.show()

#LOG scale:
###########
plt.plot(x, x**2, 'b--', x, np.exp(x),'g-')
plt.title('Log Scale')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.xscale('linear')
plt.yscale('log') #change Y scale to a log

plt.show()
```
